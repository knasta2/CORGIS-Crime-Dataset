```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("state_crime.csv")
df = df[["State", "Year", "Data.Rates.Violent.All"]].rename(columns={"Data.Rates.Violent.All": "ViolentAll"})
df
```

```python
legal_before_list = ["New York", "Washington", "Illinois", "California"]
legal_before = df.loc[df['State'].isin(legal_before_list)]
legal_before
```

```python
legal_before['year_legal'] = np.where(legal_before["State"].isin(['Illinois', 'California']), 1973, 1970)
legal_before
```

```python
legal_before.to_csv("legal_before.csv")
```

```python
easily_available_list = ["Alabama", "Texas", "Oregon", "California"]
easily_available = df.loc[df['State'].isin(easily_available_list)]
easily_available
```

```python
easily_available.to_csv("easily_available.csv")
```

```python
big_population_list = ["Florida", "Ohio", "Georgia", "Arizona", "Colorado", "Indiana"]
big_population = df.loc[df['State'].isin(big_population_list)]
big_population
```
